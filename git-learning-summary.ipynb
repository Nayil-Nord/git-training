## 3. Работа с Git

### Использованные команды:
```markdown
# Основные команды
git init               # Инициализация репозитория
git status            # Проверка статуса репозитория
git add .             # Добавление всех файлов в индекс
git commit -m "..."   # Создание коммита с сообщением
git push              # Отправка изменений на GitHub
git branch            # Просмотр веток
git checkout          # Переключение между ветками/коммитами
git log               # Просмотр истории коммитов

# Команды для работы с ветками
git checkout -b master    # Создание и переключение на новую ветку
git push -u origin master # Публикация ветки на GitHub

# Навигация по директориям
cd ..                # Переход на уровень выше
cd dirname          # Переход в директорию
ls                  # Просмотр содержимого директории

# Работа с файлами
echo "text" > file.txt   # Создание файла с текстом
rm filename             # Удаление файла
mkdir dirname          # Создание директории
```
